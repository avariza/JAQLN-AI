```python
import spacy
import pandas as pd
from nlp_utilities import clean_text
```

```python
#python -m spacy download en_core_web_sm
nlp = spacy.load("en_core_web_sm")
```

```python
data = pd.read_csv('./data.csv')
```

```python
data.head()
```

```python
df = data[['SubAgency','Description','Location','Accident','Belts','Personal_Injury','Property_Damage','Fatal',
'Commercial_License','HAZMAT','Commercial_Vehicle','Alcohol','Work_Zone','Search_Conducted',
'Search_Disposition','Search_Outcome','Search_Reason','Search_Type',
'Search_Arrest_Reason','State','VehicleType','Year','Make','Model','Color','Violation_Type',
'Article','Contributed_To_Accident','Race','Gender','Driver_City','Driver_State', 
'Arrest_Type','Day_of_Week','Hour_of_Day']]

random_sample = df.sample(frac=0.05, random_state=42).reset_index(inplace=False) 
pd.set_option('display.max_columns', None)
random_sample.head()
```

```python
random_sample[['Arrest_Category', 'Arrest_Description']] = random_sample['Arrest_Type'].apply(lambda x: pd.Series(str(x).split(" - ")))
random_sample[['Vehicle_Category', 'Vehicle_Type']] = random_sample['VehicleType'].apply(lambda x: pd.Series(str(x).split(" - ")))
```

```python
gender_mapping = {'F': 'woman', 'M': 'man', 'U': 'unidentified'}
random_sample['Gender'] = random_sample['Gender'].replace(gender_mapping)
```

```python
del random_sample['index']
random_sample.head()
```

## Normalization

```python
random_sample['Description'] = random_sample['Description'].str.lower()
random_sample['SubAgency'] = random_sample['SubAgency'].str.lower()
random_sample['Location'] = random_sample['Location'].str.lower()
random_sample['Article'] = random_sample['Article'].str.lower()
random_sample['Violation_Type'] = random_sample['Violation_Type'].str.lower()
random_sample['Search_Reason'] = random_sample['Search_Reason'].str.lower()
random_sample['Search_Type'] = random_sample['Search_Type'].str.lower()
random_sample['Search_Arrest_Reason'] = random_sample['Search_Arrest_Reason'].str.lower()
random_sample['Arrest_Description'] = random_sample['Arrest_Description'].str.lower()
random_sample['Vehicle_Type'] = random_sample['Vehicle_Type'].str.lower()
```

### Noise cleaning

```python
random_sample['Description'] = clean_text(random_sample['Description'])
random_sample['Location'] = clean_text(random_sample['Location'])
random_sample['SubAgency'] = clean_text(random_sample['SubAgency'])
```

```python
pd.set_option('display.max_columns', None)
random_sample.head(3)
```

```python
categorical_cols = random_sample.select_dtypes(include=['object']).columns

binary_columns = []
short_columns = []
long_columns = []


for col in categorical_cols:
    if random_sample[col].nunique() < 3:
        binary_columns.append(col)    
    elif random_sample[col].nunique() > 3 and random_sample[col].nunique()<11:
        short_columns.append(col)    
    else:
       long_columns.append(col)

print(binary_columns)
print(short_columns)
print(long_columns)
```
