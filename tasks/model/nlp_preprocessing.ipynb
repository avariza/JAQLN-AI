```python
import spacy
import pandas as pd
from sklearn.model_selection import train_test_split
from nlp_utilities import clean_text,tokenize_text,tokenize_text_with_numbers
```

```python
#python -m spacy download en_core_web_sm
nlp = spacy.load("en_core_web_sm")
```

```python
data = pd.read_csv('./data.csv')
data.head()
```

```python
train_df, test_and_valid_df = train_test_split(data, test_size=0.3, random_state=42)
valid_df, test_df = train_test_split(test_and_valid_df, test_size=0.5, random_state=42)

print("Train set shape:", train_df.shape)
print("Validation set shape:", valid_df.shape)
print("Test set shape:", test_df.shape)
```

```python
df = train_df[['SubAgency','Description','Location','Accident','Belts','Personal_Injury','Property_Damage','Fatal',
'Commercial_License','HAZMAT','Commercial_Vehicle','Alcohol','Work_Zone','Search_Conducted',
'Search_Disposition','Search_Outcome','Search_Reason','Search_Type',
'Search_Arrest_Reason','State','VehicleType','Year','Make','Model','Color','Violation_Type',
'Article','Contributed_To_Accident','Race','Gender','Driver_City','Driver_State', 
'Arrest_Type','Day_of_Week','Hour_of_Day']]

train_sample = df.sample(frac=0.05, random_state=42).reset_index(inplace=False) 
pd.set_option('display.max_columns', None)
print(train_sample.shape)
train_sample.head()
```

```python
train_sample[['Arrest_Category', 'Arrest_Description']] = train_sample['Arrest_Type'].apply(lambda x: pd.Series(str(x).split(" - ")))
train_sample[['Vehicle_Category', 'Vehicle_Type']] = train_sample['VehicleType'].apply(lambda x: pd.Series(str(x).split(" - ")))
```

```python
gender_mapping = {'F': 'woman', 'M': 'man', 'U': 'unidentified'}
train_sample['Gender'] = train_sample['Gender'].replace(gender_mapping)
```

```python
del train_sample['index']
train_sample.head()
```

## Normalization

```python
train_sample['Description'] = train_sample['Description'].str.lower()
train_sample['SubAgency'] = train_sample['SubAgency'].str.lower()
train_sample['Location'] = train_sample['Location'].str.lower()
train_sample['Article'] = train_sample['Article'].str.lower()
train_sample['Violation_Type'] = train_sample['Violation_Type'].str.lower()
train_sample['Search_Reason'] = train_sample['Search_Reason'].str.lower()
train_sample['Search_Type'] = train_sample['Search_Type'].str.lower()
train_sample['Search_Arrest_Reason'] = train_sample['Search_Arrest_Reason'].str.lower()
train_sample['Arrest_Description'] = train_sample['Arrest_Description'].str.lower()
train_sample['Vehicle_Type'] = train_sample['Vehicle_Type'].str.lower()
```

### Noise cleaning

```python
train_sample['Description'] = clean_text(train_sample['Description'])
train_sample['Location'] = clean_text(train_sample['Location'])
train_sample['SubAgency'] = clean_text(train_sample['SubAgency'])
```

```python
pd.set_option('display.max_columns', None)
train_sample.head(3)
```

```python
categorical_cols = train_sample.select_dtypes(include=['object']).columns

binary_columns = []
short_columns = []
long_columns = []


for col in categorical_cols:
    if train_sample[col].nunique() < 3:
        binary_columns.append(col)    
    elif train_sample[col].nunique() > 3 and train_sample[col].nunique()<11:
        short_columns.append(col)    
    else:
       long_columns.append(col)

print(binary_columns)
print(short_columns)
print(long_columns)
```
